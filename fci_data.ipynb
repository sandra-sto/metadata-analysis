### Calculate error between FCIs

```python
fci_columns = ['offsetCornerLatitude0', 'offsetCornerLatitude1', 'offsetCornerLatitude2', 'offsetCornerLatitude3', 
               'offsetCornerLongitude0', 'offsetCornerLongitude1', 'offsetCornerLongitude2', 'offsetCornerLongitude3', 'file_timestamp']
```

```python
fci_original = original_df[fci_columns]
fci_georeferenced = georeferenced_df[fci_columns]
```

```python
fci_original.set_index('file_timestamp', inplace=True)
fci_georeferenced.set_index('file_timestamp', inplace=True)

fci_original.sort_index(inplace=True)
fci_georeferenced.sort_index(inplace=True)
```

```python
fci_original.index = fci_original.index - fci_original.index[0]
```

```python
fci_original.columns = ['offset_lat_0', 'offset_lat_1', 'offset_lat_2', 'offset_lat_3',
                       'offset_lon_0', 'offset_lon_1', 'offset_lon_2', 'offset_lon_3']
```

```python
fci_georeferenced.columns = ['geo_offset_lat_0', 'geo_offset_lat_1', 'geo_offset_lat_2', 'geo_offset_lat_3',
                       'geo_offset_lon_0', 'geo_offset_lon_1', 'geo_offset_lon_2', 'geo_offset_lon_3']
```

```python
fci_georeferenced.iloc[30:60]
```

```python
#joined_fci = fci_original.join(fci_georeferenced)
merged_fci = pd.merge_asof(fci_original, fci_georeferenced, direction='nearest', left_index=True, right_index=True)
```

```python
diff_fci = pd.DataFrame(index = merged_fci.index, columns=['diff_lat_0', 'diff_lat_1', 'diff_lat_2', 'diff_lat_3',
                       'diff_lon_0', 'diff_lon_1', 'diff_lon_2', 'diff_lon_3'])
```

```python
for i in range(8):
    diff_fci.iloc[:,i] =  merged_fci.iloc[:,i] - merged_fci.iloc[:,i+8]
```
